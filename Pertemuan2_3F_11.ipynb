```python
#2 Menampilkan Gambar
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/PCVK25_3F_11/Gambar/KTM lama.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#3 Menampilkan citra Grayscale:

img_gray = cv2.imread("/content/drive/MyDrive/PCVK25_3F_11/Gambar/KTM lama.jpg", cv2.IMREAD_GRAYSCALE)

#Tanpa cmap
plt.subplot(1, 3, 1)
plt.imshow(img_gray)
plt.title('Tanpa cmap')

#Dengan cmap
plt.subplot(1, 3, 2)
plt.imshow(img_gray, cmap='gray')
plt.title('Dengan cmap')

#Mencoba dengan warna lain
plt.subplot(1, 3, 3)
plt.imshow(img_gray, cmap='magma')
plt.title('dengan cmap magma')

plt.show()
```

```python
#4 Melakukan Rezizing

img = cv2.imread("/content/drive/MyDrive/PCVK25_3F_11/Gambar/KTM lama.jpg")
imgRGB= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imgRGMsmall = cv2.resize(imgRGB,(400, 250))
plt.imshow(imgRGMsmall)
```

```python
#5 Melakukan Flipping

imgBalik = cv2.flip(imgRGB, 1)
plt.imshow(imgBalik)

#Menampilkan kanvas yang lebih besar
fig=plt.figure(figsize=(10, 10))
ax= fig.add_subplot(111)
ax.imshow(imgBalik)
```

```python
#6 Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image dengan tipe data int16.
import numpy as np

black_image = np.zeros(shape=(512, 512, 3), dtype=np.int16)
plt.imshow(black_image)

#Membuat Kotak Hijau
#perhatikan koordinat titik2 pt1 dan pt2
cv2.rectangle(black_image, pt1=(384, 10), pt2=(510, 150), color=(0, 255, 0), thickness=10)
plt.imshow(black_image)
```

```python
#menambahkan bentuk persegi sesuai koordinat pt1 dan pt2
cv2.rectangle(black_image, pt1=(200, 200), pt2=(300, 300), color=(0, 0, 255), thickness=15)
plt.imshow(black_image)
```

```python
#menambahkan bentuk lingkaran sesuai radius
cv2.circle(black_image, center=(100, 100), radius=50, color=(255, 0, 0), thickness=8)
plt.imshow(black_image)
```

```python
#penambahan garis sesuai koordinat pt1 dan pt2
cv2.line(black_image, pt1=(0, 0), pt2=(512, 512), color=(255, 255, 255), thickness=5)
plt.imshow(black_image)
```

```python
#Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah ditentukan

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(black_image, text='Hello', org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 0), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(black_image)
```

```python
#Pembuatan black image dengan tipe data int32
black_img2 = np.zeros(shape=(512, 512, 3), dtype=np.int32)
plt.imshow(black_img2)
```

```python
#inisialisasi NumPy array dengan tipe data int32
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)
vertices
```

```python
#Mereshape Array
pts = vertices.reshape((-1, 1, 2))
pts
```

```python
#Penambahan polyline pada black image kedua yang telah dibuat.
cv2.polylines(black_img2, [pts], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(black_img2)
```

# ***TUGAS PRAKTIKUM D3***

```python
#Import Library
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import random

#Load Image
image = cv.imread("/content/drive/MyDrive/PCVK25_3F_11/Gambar/KTM lama.jpg")
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # konversi BGR ke RGB
```

```python
#1 Dengan menggunakan figsize, perhatikan apakah ukuran image pixelnya juga berubah?
plt.figure(figsize=(10,5))
plt.imshow(image)
plt.title("Ukuran Figure Lebih Besar - Pixel Tetap Sama")
plt.show()
```

```python
#2 Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

RB = image.copy()
RB[:,:,1] = 0   # menghilangkan channel Green
GB = image.copy()
GB[:,:,0] = 0   # menghilangkan channel Red

plt.subplot(1,2,1)
plt.imshow(RB)
plt.title("Red-Blue Channel")

plt.subplot(1,2,2)
plt.imshow(GB)
plt.title("Green-Blue Channel")
plt.show()
```

```python
# 3. Tampilkan image baris 20-115, kolom 25-120
crop1 = image[20:115, 25:120, :]
plt.imshow(crop1)
plt.title("Cropped (baris 20-115, kolom 25-120)")
plt.show()
```

```python
# 4. Tampilkan image baris ke 5-30, semua kolom, channel Red saja
red_channel = image[5:30, :, 0]
plt.imshow(red_channel, cmap="Reds")
plt.title("Red Channel (baris 5-30, semua kolom)")
plt.show()
```

```python
# 5. Buat 5 kotak acak berbagai ukuran dan warna
image_kotak = image.copy()
t, l, _ = image.shape
for _ in range(5):
    x1, y1 = random.randint(0,l-50), random.randint(0,t-50)
    x2, y2 = random.randint(x1+20, min(x1+100,l)), random.randint(y1+20, min(y1+100,t))
    color = (random.randint(0,255), random.randint(0,255), random.randint(0,255))
    cv.rectangle(image_kotak, (x1,y1), (x2,y2), color, -1)

plt.imshow(image_kotak)
plt.title("Image dengan 5 Kotak Acak")
plt.show()
```

```python
# 6. Tampilkan image dengan posisi terbalik (flip)
flip_vertical = cv.flip(image, 0)    # flip atas-bawah
flip_horizontal = cv.flip(image, 1)  # flip kiri-kanan

plt.subplot(1,2,1)
plt.imshow(flip_vertical)
plt.title("Flip Vertikal")

plt.subplot(1,2,2)
plt.imshow(flip_horizontal)
plt.title("Flip Horizontal")
plt.show()
```

```python
#7.Buat rectangle dan circle pada bagian wajah dari image foto s anda saat beraktifitas (bukan pasfoto).
image_face = cv.imread("/content/drive/MyDrive/PCVK25_3F_11/Gambar/FALI HIMA.jpg")
image_face = cv.cvtColor(image_face, cv.COLOR_BGR2RGB)

# Tentukan koordinat wajah (x,y,w,h) secara manual
x, y, w, h =  300, 500, 300, 350
cv.rectangle(image_face, (x, y), (x+w, y+h), (255, 0, 0), 3)  # kotak
cv.circle(image_face, (x+w//2, y+h//2), min(w,h)//3, (0,255,0), 3)  # lingkaran

plt.imshow(image_face)
plt.title("Rectangle dan Circle pada Wajah")
plt.show()
```

```python
#8 Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/
# peppers!

image_kitten = cv.imread("/content/drive/MyDrive/PCVK25_3F_11/Gambar/kitten01.jpg")
image_kitten= cv.cvtColor(image_kitten, cv.COLOR_BGR2RGB)

# Ambil channel Blue
B_img = image_kitten.copy()
B_img[:,:,0] = 0   # matikan R
B_img[:,:,1] = 0   # matikan G

h, w, _ = B_img.shape
x1, y1 = 0, h-100
x2, y2 = 100, h

cv.rectangle(B_img, (x1,y1), (x2,y2), (255,255,255), 3)

plt.imshow(B_img)
plt.title("Rectangle pada Sudut Kiri Bawah Channel B (3 channel)")
plt.show()
```
